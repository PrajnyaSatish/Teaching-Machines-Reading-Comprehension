```python
from matplotlib import pyplot as plt
import json

from functools import reduce
import re
import numpy as np
import nltk
import json
from pprint import pprint as pp
```

```python
def tokenize(sent):
    """Returns the tokens of a sequece"""
    tokens = nltk.word_tokenize(sent)
    tokens = [w.lower() for w in tokens]
    return tokens
```

```python
with open('data/SciQ dataset/train.json', 'r') as rf:
    train = json.load(rf)
with open('data/SciQ dataset/test.json', 'r') as rf:
    test = json.load(rf)
with open('data/SciQ dataset/valid.json', 'r') as rf:
    valid = json.load(rf)
```

```python
all_question_lengths = []
all_support_lengths = []

for sample in train+test+valid:
    question_length = len(tokenize(sample['question']))
    support_length = len(tokenize(sample['support']))
    all_question_lengths.append(question_length)
    all_support_lengths.append(support_length)
```

```python
plt.hist(all_support_lengths)
plt.title("Histogram plot of length of Support")
plt.xlabel("Length Ranges")
plt.ylabel("Frequency")
```

```python
plt.hist(all_question_lengths)
plt.title("Histogram plot of length of Question")
plt.xlabel("Length Ranges")
plt.ylabel("Frequency")
```
